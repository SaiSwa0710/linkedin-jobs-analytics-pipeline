```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from tqdm import tqdm
import os
import csv
import torch
import numpy as np
import ast
from tqdm import tqdm
import ast
from sklearn.decomposition import PCA
import pyarrow as pa
import pyarrow.parquet as pq
import math
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
model = SentenceTransformer("paraphrase-MiniLM-L3-v2", device=device)
```

```python
print("Running on:", device)
```

```python
INPUT_FILE = "job_summary.csv"
OUTPUT_FILE = "job_summary_embeddings.csv"
```

```python
completed_links = set()
open(OUTPUT_FILE, "w").write("job_link,embed\n")
print("Starting fresh.")
```

```python
chunk_size = 5000
processed = len(completed_links)

for chunk in pd.read_csv(INPUT_FILE, chunksize=chunk_size):

    chunk = chunk[~chunk["job_link"].isin(completed_links)]
    if len(chunk) == 0:
        continue

    summaries = chunk["job_summary"].fillna("").astype(str).tolist()

    # GPU encoding
    embeddings = model.encode(
        summaries,
        batch_size=256,
        convert_to_numpy=True,
        show_progress_bar=True
    )

    embed_str = [np.array(e).tolist() for e in embeddings]

    out_df = pd.DataFrame({
        "job_link": chunk["job_link"].values,
        "embed": embed_str
    })

    out_df.to_csv(OUTPUT_FILE, mode="a", index=False, header=False)

    processed += len(chunk)
    print(f"Processed total: {processed}")
```

```python
sample_rows = 50000

sample = pd.read_csv(
    "job_summary_embeddings.csv",
    nrows=sample_rows,
    converters={"embed": ast.literal_eval}
)

X = sample["embed"].tolist()

pca = PCA(n_components=32)
pca.fit(X)

print("PCA fitted successfully!")
```

```python



input_path = "job_summary_embeddings.csv"
output_path = "job_summary_embeddings_32d.csv"

chunk_size = 10_000

pd.DataFrame(columns=["job_link"] + [f"d{i}" for i in range(32)]).to_csv(
    output_path, index=False
)

for chunk in tqdm(pd.read_csv(
        input_path,
        chunksize=chunk_size,
        converters={"embed": ast.literal_eval}
    ),
    desc="Compressing to 32-D"
):
    embeds = np.array(chunk["embed"].tolist())          # (N, 384)
    embeds_32 = pca.transform(embeds)                   # (N, 32)

    out = pd.DataFrame({
        "job_link": chunk["job_link"],
        **{f"d{i}": embeds_32[:, i] for i in range(32)}
    })

    out.to_csv(output_path, mode="a", header=False, index=False)

print("Done  new 32-D compressed file saved!")
```

```python

csv_file = "job_summary_embeddings_32d.csv"

df_head = pd.read_csv(csv_file, nrows=0)
columns = df_head.columns.tolist()

print("Columns in the CSV file:", columns)
```

```python


csv_file = "job_summary_embeddings_32d.csv"
parquet_file = "job_summary_embeddings_32d_list.parquet"

chunk_size = 50000 
columns_to_keep = ["job_link"] + [f"d{i}" for i in range(32)]

total_rows = sum(1 for _ in open(csv_file)) - 1
total_chunks = math.ceil(total_rows / chunk_size)

writer = None

# Read CSV in chunks
for chunk in tqdm(pd.read_csv(csv_file, usecols=columns_to_keep, chunksize=chunk_size),
                  total=total_chunks,
                  desc="Converting CSV  Parquet",
                  unit="chunk"):

    chunk["element"] = chunk[[f"d{i}" for i in range(32)]].values.tolist()
    chunk = chunk[["job_link", "element"]]

    table = pa.Table.from_pandas(chunk)

    if writer is None:
        writer = pq.ParquetWriter(parquet_file, table.schema)
    writer.write_table(table)

    del chunk, table

if writer:
    writer.close()

print("Done! CSV converted to Parquet with single list column 'element':", parquet_file)
```
